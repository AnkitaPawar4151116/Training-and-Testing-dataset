```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
#read dataset
ds=pd.read_csv('customer_data.csv')
ds
```

```python
#encode Name column becuase it is categorical data (OneHotEncoder)
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output=False)
encode_data=ohe.fit_transform(ds[['Name']])
encode_data
```

```python
#giving names to new columns
column_name=ohe.get_feature_names_out(['Name'])
encode_df=pd.DataFrame(encode_data,columns=column_name)
encode_df
```

```python
# concat to OG dataset
ds=pd.concat([ds,encode_df],axis=1)
ds
```

```python
#drop original old column
ds=ds.drop(columns=['Name'])
ds
```

```python
##encode Gender column becuase it is categorical data (LabelEncoder use when bidimentional data i.e true/false,yes/no etc )
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
ds['Gender']=le.fit_transform(ds['Gender'])
ds
```

```python
##encode Purchased column becuase it is categorical data (LabelEncoder use when bidimentional data i.e true/false,yes/no etc )
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ds['Purchased']=le.fit_transform(ds['Purchased'])
ds
```

```python
#encode Name column becuase it is categorical data (OneHotEncoder)
from sklearn.preprocessing import OneHotEncoder
ohe= OneHotEncoder(sparse_output=False)
encode_data=ohe.fit_transform(ds[['City']])
encode_data
```

```python
#giving names to new columns
column_name=ohe.get_feature_names_out(['City'])
encode_df=pd.DataFrame(encode_data,columns=column_name)
encode_df
```

```python
# concat to OG dataset
ds=pd.concat([ds,encode_df],axis=1)
ds
```

```python
#drop original old column
ds=ds.drop(columns=['City'])
ds
```
